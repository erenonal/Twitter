# Twitter Text Mining

```python
df = hashtag_df(tweetler)

#buyuk-kucuk donusumu
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#noktalama i≈üaretleri
df['text'] = df['text'].str.replace('[^\w\s]','')

#sayƒ±lar
df['text'] = df['text'].str.replace('\d','')

#stopwords
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sw))

#lemmi
from textblob import Word
#nltk.download('wordnet')
df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 

#noktalama i≈üaretleri
df['text'] = df['text'].str.replace('rt','')
```

```python
from textblob import TextBlob
```

```python
def sentiment_skorla(df):
    #dataframe'in i√ßinden tweetleri aldƒ±k
    text = df['text']
    #tweetleri dola≈ütƒ±k
    for i in range(0, len(text)):
        #alƒ±nan tweeti textB'ye atadƒ±k
        textB = TextBlob(text[i])
        #alƒ±nan tweet'e g√∂re sentiment.polarity fonksiyonu ile skor verdik
        sentiment_skoru = textB.sentiment.polarity
        #dataframe'in ilgili s√ºtununa sentiment skorunu ekledik
        df.at[i, 'sentiment_skoru'] = sentiment_skoru
        #ilgili s√ºtuna sentiment skoruna g√∂re duygu sƒ±nƒ±fƒ± ekledik
        if sentiment_skoru <0.00:
            duygu_sinifi = 'Negatif'
            df.at[i, 'duygu_sinifi'] = duygu_sinifi
        elif sentiment_skoru >0.00:
            duygu_sinifi = 'Pozitif'
            df.at[i, 'duygu_sinifi'] = duygu_sinifi
        else:
            duygu_sinifi = 'Nort'
            df.at[i, 'duygu_sinifi'] = duygu_sinifi
    return df
```
